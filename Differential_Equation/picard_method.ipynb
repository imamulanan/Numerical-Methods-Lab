```python
import numpy as np
import matplotlib.pyplot as plt

# Picard Approximation গুলো (Y1, Y2, Y3)
def Y1(x):
    return 1 + x + (x**2) / 2

def Y2(x):
    return 1 + x + (x**2) / 2 + (x**3) / 3 + (x**4) / 8

def Y3(x):
    return 1 + x + (x**2) / 2 + (x**3) / 3 + (x**4) / 8 + (x**5) / 15 + (x**6) / 48

# Picard Method (সহজভাবে)
def picard(f, x0, h=0.1, n=10):
    x = np.arange(x0, x0 + n*h, h)
    y = np.array([f(i) for i in x])
    return x, y

# Y3 দিয়ে গণনা করা
x, y = picard(Y3, 0, 0.1)

# ফলাফল প্লট করা
plt.plot(x, y, color='purple', label='Picard Approximation (Y3)')
plt.scatter(x[1], y[1], color='red', label=f'Point at x={x[1]:.1f}')
plt.scatter(x[2], y[2], color='green', label=f'Point at x={x[2]:.1f}')
plt.title("Picard's Method Approximation")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
