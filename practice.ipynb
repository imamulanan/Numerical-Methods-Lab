```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
x=2*np.random.rand(100,1)
y=4+3*x+np.random.randn(100,1)
```

```python
def sgd(x,y,m=0,b=0,learning_rate=0.01,epochs=1000):
    n=len(y)
    for _ in range(epochs):
        for i in range(n):
            xi=x[i:i+1]
            yi=y[i:i+1]
            y_pred=m*xi+b
            dm = -2*xi*(yi-y_pred)
            db = -2*(yi-y_pred)
            m -= learning_rate*dm
            b -= learning_rate*db
    return m,b
```

```python
plt.scatter(x,y)

m,b = gd(x,y)
plt.plot(x,m*x+b,color='red')

plt.xlabel("X")
plt.ylabel("Y")

plt.title("Gradient Descent")
plt.show()
```
