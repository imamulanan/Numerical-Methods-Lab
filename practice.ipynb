```python
import numpy as np
import matplotlib.pyplot as plt

def f(x,y):
    return x+y

def rk4(f,x0,x1,y0,h):
    n = int((x1-x0)/h)
    x=np.linspace(x0,x1,n+1)
    y=np.zeros(n+1)
    y[0]=y0

    for i in range(n):
        k1 = h*f(x[i],y[i])
        k2 = h*f(x[i]+h/2, y[i]+k1/2)
        k3 = h*f(x[i]+h/2, y[i]+k2/2)
        k4 = h*f(x[i]+h,y[i]+k3)

        y[i+1] = y[i] + (k1 + 2*(k2+k3) + k4)/6

    return x,y

x0,y0 = 0,1
x1,h = 10,0.1

x,y = rk4(f,x0,x1,y0,h)


plt.plot(x,y,'r',label = 'Runge Kutta 4th Order')
plt.xlabel('x')
plt.ylabel('y')
plt.title("RK4 Method")
plt.legend()
plt.grid(True)
plt.show()
```
