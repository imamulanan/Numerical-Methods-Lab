```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])

m,c = np.polyfit(x,y,1)

y_pred = m*x +c

plt.scatter(x,y,color='blue',label='Data Points')
plt.plot(x,y_pred,color='red')
plt.legend()
plt.show()

print(m*6 + c)  # Predicting for x=6    
```
