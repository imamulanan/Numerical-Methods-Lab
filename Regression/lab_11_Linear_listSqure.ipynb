```python
import numpy as np
import matplotlib.pyplot as plt

# Data
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])

# Line fit (y = mx + c)
m, c = np.polyfit(x, y, 1)

# Prediction line
y_pred = m * x + c

# Plot
plt.scatter(x, y, color="blue")
plt.plot(x, y_pred, color="red", label=f"y={m:.2f}x+{c:.2f}")
plt.legend()
plt.show()

# Predict for x = 6
print(f"Predicted value for x=6: {m*6 + c:.2f}")
```
