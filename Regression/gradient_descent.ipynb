```python
import numpy as np
import matplotlib.pyplot as plt

# Random data
np.random.seed(0)

# x = 2 * np.random.rand(100, 1): 0 theke 2 er moddhe 100 ta random x value generate kore (column vector).
x = 2 * np.random.rand(100, 1)

# y = 4 + 3 * x + np.random.randn(100, 1): y value create kore, jeikhane y = 4 + 3x + noise (noise holo normally distributed random value, real-life data-r moto).
y = 4 + 3*x + np.random.randn(100, 1)

# Gradient Descent function
def gradient_descent(x, y, lr=0.01, epochs=10000):
    m, b = 0, 0  # initial slope and intercept
    n = len(y)
    for _ in range(epochs):
        y_pred = m*x + b
        m -= lr * (-2/n) * np.sum(x * (y - y_pred))
        b -= lr * (-2/n) * np.sum(y - y_pred)
    return m, b

# Fit model
m, b = gradient_descent(x, y)

# Plot
plt.scatter(x, y, label='Data')
plt.plot(x, m*x + b, color='red', label=f'Line: y={m:.2f}x+{b:.2f}')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Gradient Descent Linear Regression')
plt.legend()
plt.show()
```
