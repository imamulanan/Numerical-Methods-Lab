```python
# Lagrange Interpolation Function
def lagrange_interpolation(data, x_n):
    result = 0
    n = len(data)
    
    for i in range(n):
        xi, yi = data[i]  # xi and yi are the data points
        term = yi
        
        # Loop to calculate the product for the Lagrange basis polynomial
        for j in range(n):
            if j != i:  # Skip when i == j
                xj, _ = data[j]
                term *= (x_n - xj) / (xi - xj)
        
        result += term  # Add the result for each term
        
    return result

# Example Data Points
data = [(5, 12), (6, 13), (9, 14), (11, 16)]
x_n = 10  # Value to interpolate at

# Call the function and get the interpolated result
y_n = lagrange_interpolation(data, x_n)
print(f"Interpolated value at x = {x_n}: {y_n}")
```
